## Profit vs Sales

```python
sns.set(font_scale=1.5)  
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(data['Sales'], data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

## Profit vs Discount

```python
sns.lineplot(x="Discount", y="Profit",label = "Profit",data=data)
plt.legend()
plt.show()
```

###### Observation : We can see there is no correlation between profit and discount

## Profit vs Quantity

```python
sns.lineplot(x="Quantity", y="Profit",label = "Profit",data=data)
plt.legend()
plt.show()
```

## Visualizing Profit and Sales across different Segment

```python
data.groupby("Segment")[['Profit', 'Sales']].sum().plot.bar(color = ["brown", "orange"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

##### Observation : Sales and Profit were highest in consumer segment and least in Home office segment

## Visualizing segment wise sales distribution in each region

```python
plt.figure(figsize=(12,8))
plt.title("Distribution of Segment wise sales in each region " , fontsize=15)
sns.barplot(x="Region",y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='rainbow')
plt.xlabel('Region', fontsize=15)
plt.tick_params(labelsize=12)
plt.show()
```

##### Observation : Segment wise sales are almost same in every region.

## Visualizing Profit and sales across different regions

```python
data.groupby("Region")[['Profit', 'Sales']].sum().plot.bar(color = ["red", "orange"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

###### Observation : Western region has the highest sales and profit whereas southern region has the least sales , but a good profit wrt its sales.

## Visualizing profit and sales across states

```python
csp = data.groupby("State")[["Sales", "Profit"]].sum().sort_values(by = "Sales", ascending = False)
csp[:].plot.bar(color = ["orange", "blue"],figsize= (18,8))
plt.title("Profit/Loss & Sales across States")
plt.xlabel("States")
plt.ylabel("Profit/Loss & Sales")
plt.show()
```

##### Observation : States like California and New York have high sales with a good profit whereas staes like Texas, Pennsylvania, Ohio face huge losses

```python
#top 10 states with high sales
top_states= data['State'].value_counts().nlargest(10)
top_states
```

## Visualizing Profit and Sales across different Categories

```python
data.groupby("Category")[['Profit', 'Sales']].sum().plot.bar(color = ["brown", "red"],figsize= (8,5))
plt.ylabel("Profit /loss and Sales")
plt.show()
```

#### Observation:    

1. Technology and office supplies category have high sales with good profit.
2. Furniture category has least profit.

## Visualizing profit and sales across various Sub-Categories

```python
scsp=data.groupby("Sub-Category")[["Sales","Profit"]].mean().sort_values(by = "Sales", ascending = False)
scsp.plot.bar(color=['brown','red'],figsize= (18,8))
plt.title("Profit and Sales across various Sub-Categories",fontsize=20)
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel("Profit/Loss & Sales")
plt.tick_params(labelsize=14)
plt.show()
```

## Note:

1. Maximum number of buyers are from California and New York.
2. Maximum number of buyers prefer standard class ship mode.

5. States like California and New York have high sales with a good profit whereas states like Texas, Pennsylvania, Ohio face huge losses.
